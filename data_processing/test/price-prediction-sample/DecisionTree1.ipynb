```
import pandas as pd
data = pd.read_csv('melb_data.csv')
data
```

```
## FInd all the numerical columns in data
data_predictors = list(data.dtypes[data.dtypes == "float"].index)
data_predictors = data_predictors[:-1]
data_new = data[data_predictors].dropna()
X = data_new.drop('Price', axis=1)
y = data_new.Price
print(data_predictors)
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```
def accuracy(predict,real):
  arr = []
  for i in range(len(real)):
    arr.append((abs(predict[i]-real[i]))/real[i])
  print((1-(sum(arr)/len(arr)))*100)
  return
```

```
y_pred = model.predict(X_test)
accuracy(y_pred,y_test.values)
```
